```python
#21
def my_function_21():
  list_1 = [1, 2, 3]
  list_2 = [4, 5, 6]
  if len(list_1) != len(list_2):
    return "lists must be of the same length"
  product = 0
  for i in range(len(list_1)):
    product += list_1[i] * list_2[i]
  return product

print(my_function_21())
```

```python
#22
def dot_product(u,v):
  return sum(x * y for x, y in zip(u,v))
def scalar_multiply(scalar, vector):
  return [scalar * x for x in vector]
def projection(a,b):
  dot_ab = dot_product(a,b)
  dot_bb = dot_product(b,b)
  return "cannot project onto a zero vector"
  scalar = dot_ab / dot_bb
  return scalar_multiply(scalar, b)
a = [3, 4]
b = [1, 2]
print("projection of a onto b:", projection(a,b))
```

```python
#25
def my_function_25(data):
  data = sorted(data)
  n = len(data)

  q1 = data[n//4]
  q3 = data[(3*n)//4]
  iqr = q3 - q1

  low = q1 - 1.5 * iqr
  high = q3 + 1.5 * iqr

  return [x for x in data if x < low or x > high]
data = [ 10, 12, 13, 14, 15, 100]
print(my_function_25(data))
```

```python
#28
import math
def my_function_28(vector):
  l1 = sum(abs(x) for x in vector)
  l2 = math.sqrt(sum(x**2 for x in vector))
  return l1, l2
v = [3, -4, 5]
l1, l2 = my_function_28(v)
print("l1 norm:", l1)
print("l2 norm:", l2)
```
